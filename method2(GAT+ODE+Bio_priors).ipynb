```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GATConv, GCNConv
from torchdiffeq import odeint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import re
from scipy.stats import pearsonr, spearmanr
from sklearn.metrics import f1_score, precision_recall_curve
import networkx as nx
from google.colab import files as colab_files
import warnings
warnings.filterwarnings('ignore')

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
```

```python
# ============================================================================
# BLOCK 2: Data Loading and Preprocessing
# ============================================================================

# @title Load Data Files
def get_files():
    """Load Witvliet connectome files and coordinate data"""
    # Look for files
    extensions = ['*.csv', '*.xlsx', '*.xls']
    found_files = []
    for ext in extensions:
        found_files.extend(glob.glob(ext))

    # Filter for Witvliet files
    witvliet_files = [f for f in found_files if "witvliet" in f.lower()]
    coord_files = [f for f in found_files if "atlas" in f.lower() or "coord" in f.lower()]

    # Upload if missing
    if len(witvliet_files) < 8:
        print(f"\nFound only {len(witvliet_files)}/8 connectome files.")
        print("Please upload all 8 Witvliet files and the coordinate CSV.")
        uploaded = colab_files.upload()

        found_files = []
        for ext in extensions:
            found_files.extend(glob.glob(ext))
        witvliet_files = [f for f in found_files if "witvliet" in f.lower()]
        coord_files = [f for f in found_files if "atlas" in f.lower() or "coord" in f.lower()]

    print(f"\nFound {len(witvliet_files)} connectome files")
    print(f"Found {len(coord_files)} coordinate files")

    return sorted(witvliet_files), coord_files[0] if coord_files else None

def parse_stage(filename):
    """Extract developmental stage from filename"""
    name = filename.lower()

    # Explicit names
    if 'l1' in name: return 0.0
    if 'l2' in name: return 1.0
    if 'l3' in name: return 2.0
    if 'adult' in name: return 3.0

    # Fallback: Extract number
    match = re.search(r'witvliet_2020_(\d)', name)
    if match:
        num = int(match.group(1))
        if 1 <= num <= 4: return 0.0  # L1
        if num == 5: return 1.0       # L2
        if num == 6: return 2.0       # L3
        if num >= 7: return 3.0       # Adult

    return None

def read_file(fpath):
    """Read CSV or Excel file"""
    if fpath.endswith('.xlsx') or fpath.endswith('.xls'):
        return pd.read_excel(fpath)
    else:
        return pd.read_csv(fpath)

def load_coordinates(coord_file):
    """Load neuron spatial coordinates"""
    if coord_file is None:
        print("Warning: No coordinate file found, using random coordinates")
        return None

    df = pd.read_csv(coord_file, header=None, names=['neuron', 'x', 'y', 'z'])
    coord_dict = {}
    for _, row in df.iterrows():
        neuron_name = str(row['neuron']).strip()
        coord_dict[neuron_name] = np.array([row['x'], row['y'], row['z']])

    print(f"Loaded coordinates for {len(coord_dict)} neurons")
    return coord_dict

def classify_cell_type(neuron_name):
    """
    Classify neurons by type based on naming conventions and known types
    Returns: 0=sensory, 1=inter, 2=motor, 3=modulatory
    """
    name = neuron_name.upper()

    # Sensory neurons (amphid, labial, etc.)
    sensory_prefixes = ['ASE', 'ASG', 'ASH', 'ASI', 'ASJ', 'ASK', 'AWA', 'AWB', 'AWC',
                       'AFD', 'ADF', 'ADL', 'BAG', 'FLP', 'IL1', 'IL2', 'OLL', 'OLQ',
                       'PHA', 'PHB', 'PHC', 'PLM', 'PLN', 'PVM', 'PVD', 'AQR', 'PQR',
                       'URX', 'URY', 'ADE', 'PDE', 'CEP']

    # Motor neurons
    motor_prefixes = ['DA', 'DB', 'DD', 'VD', 'VA', 'VB', 'AS', 'VC', 'VE',
                     'RM', 'SMD', 'SMB', 'URA', 'SAA', 'SAB', 'SIA', 'SIB',
                     'HSN', 'PVN', 'PVW', 'PVX', 'PVY', 'PVZ']

    # Modulatory neurons (serotonin, dopamine, octopamine)
    modulatory_neurons = ['AIM', 'RIC', 'NSM', 'HSN', 'ADE', 'PDE', 'CEP']

    # Muscles
    if name.startswith('M') or name.startswith('BWM') or name.startswith('PM'):
        return 4  # Muscle

    # Check modulatory first (highest specificity)
    for mod in modulatory_neurons:
        if name.startswith(mod):
            return 3

    # Check motor
    for mot in motor_prefixes:
        if name.startswith(mot):
            return 2

    # Check sensory
    for sens in sensory_prefixes:
        if name.startswith(sens):
            return 0

    # Default to interneuron
    return 1

def compute_physical_contact_from_coords(coord_dict, neuron_map, contact_threshold=50.0):
    """
    Estimate physical contact areas from 3D coordinates
    Neurons closer than threshold are assumed to have contact
    Contact strength inversely proportional to distance
    """
    num_nodes = len(neuron_map)
    contact_matrix = np.zeros((num_nodes, num_nodes))

    id_to_name = {idx: name for name, idx in neuron_map.items()}

    for i in range(num_nodes):
        for j in range(i+1, num_nodes):
            name_i = id_to_name[i]
            name_j = id_to_name[j]

            if name_i in coord_dict and name_j in coord_dict:
                coord_i = coord_dict[name_i]
                coord_j = coord_dict[name_j]

                # Euclidean distance
                dist = np.linalg.norm(coord_i - coord_j)

                # Contact strength: inverse of distance (capped)
                if dist < contact_threshold:
                    contact_strength = (contact_threshold - dist) / contact_threshold
                    contact_matrix[i, j] = contact_strength
                    contact_matrix[j, i] = contact_strength

    return contact_matrix

def process_data(file_list, coord_file=None):
    """
    Process all connectome files and build comprehensive dataset
    Returns:
        - data_by_time: Dict of adjacency matrices by time and replicate
        - neuron_map: Dict mapping neuron names to indices
        - num_nodes: Total number of unique neurons
        - coord_dict: Spatial coordinates
        - contact_matrix: Estimated physical contacts
        - cell_type_dict: Cell type classifications
    """
    print("\n" + "="*60)
    print("PROCESSING CONNECTOME DATA")
    print("="*60)

    # Load coordinates
    coord_dict = load_coordinates(coord_file)

    # Build neuron index
    all_neurons = set()
    valid_files = []

    print("\nBuilding neuron index...")
    for fpath in file_list:
        try:
            df = read_file(fpath)

            # Standardize columns
            cols = [c.lower() for c in df.columns]
            df.columns = cols

            rename_map = {}
            for c in cols:
                if 'pre' in c: rename_map[c] = 'pre'
                elif 'post' in c: rename_map[c] = 'post'
                elif 'syn' in c or 'weight' in c: rename_map[c] = 'synapses'
            df.rename(columns=rename_map, inplace=True)

            if 'pre' in df.columns and 'post' in df.columns:
                all_neurons.update(df['pre'].astype(str).unique())
                all_neurons.update(df['post'].astype(str).unique())

                t = parse_stage(fpath)
                if t is not None:
                    valid_files.append({'t': t, 'df': df, 'file': fpath})
                    print(f"✓ {os.path.basename(fpath)} -> Time: {t}")
        except Exception as e:
            print(f"✗ Error reading {fpath}: {e}")

    # Create neuron mapping
    neuron_map = {name: i for i, name in enumerate(sorted(list(all_neurons)))}
    num_nodes = len(neuron_map)
    id_to_name = {idx: name for name, idx in neuron_map.items()}

    print(f"\n{'Total Unique Neurons:':<30} {num_nodes}")
    print(f"{'Valid Connectome Files:':<30} {len(valid_files)}")

    # Classify cell types
    cell_type_dict = {}
    type_counts = {0: 0, 1: 0, 2: 0, 3: 0, 4: 0}
    type_names = {0: 'Sensory', 1: 'Inter', 2: 'Motor', 3: 'Modulatory', 4: 'Muscle'}

    for name, idx in neuron_map.items():
        cell_type = classify_cell_type(name)
        cell_type_dict[idx] = cell_type
        type_counts[cell_type] += 1

    print(f"\n{'Cell Type Distribution:':}")
    for type_id, count in type_counts.items():
        print(f"  {type_names[type_id]:<15} {count:>4}")

    # Compute physical contact matrix from coordinates
    if coord_dict is not None:
        print("\nComputing physical contact matrix from coordinates...")
        contact_matrix = compute_physical_contact_from_coords(coord_dict, neuron_map)
        contact_coverage = np.sum(contact_matrix > 0) / (num_nodes * num_nodes) * 100
        print(f"  Contact coverage: {contact_coverage:.2f}% of neuron pairs")
    else:
        print("\nWarning: No coordinates available, contact matrix will be zero")
        contact_matrix = np.zeros((num_nodes, num_nodes))

    # Convert to adjacency matrices by time and replicate
    data_by_time = {}

    print("\nBuilding adjacency matrices...")
    for item in valid_files:
        t = item['t']
        df = item['df']

        # Create adjacency matrix
        adj = np.zeros((num_nodes, num_nodes), dtype=np.float32)

        for _, row in df.iterrows():
            pre, post = str(row['pre']), str(row['post'])
            if pre in neuron_map and post in neuron_map:
                src, dst = neuron_map[pre], neuron_map[post]

                # Get synapse count (use raw counts, not log-normalized)
                val = row['synapses'] if 'synapses' in row else 1.0
                try:
                    w = float(val)
                except:
                    w = 1.0

                adj[src, dst] += w

        if t not in data_by_time:
            data_by_time[t] = []
        data_by_time[t].append(torch.tensor(adj, dtype=torch.float32))

    # Print statistics
    print(f"\n{'Time Point Statistics:':}")
    for t in sorted(data_by_time.keys()):
        n_reps = len(data_by_time[t])
        avg_synapses = sum([mat.sum().item() for mat in data_by_time[t]]) / n_reps
        avg_connections = sum([(mat > 0).sum().item() for mat in data_by_time[t]]) / n_reps
        print(f"  Time {t:.1f}: {n_reps} replicates, "
              f"{avg_synapses:.0f} synapses, {avg_connections:.0f} connections")

    return data_by_time, neuron_map, num_nodes, coord_dict, contact_matrix, cell_type_dict

# Load all data
files_list, coord_file = get_files()
data_map, neuron_map, NUM_NODES, coord_dict, contact_matrix, cell_type_dict = \
    process_data(files_list, coord_file)

print("\n" + "="*60)
print("DATA LOADING COMPLETE")
print("="*60)
```

```python
# ============================================================================
# BLOCK 3: Build Rich Node Features and Prepare Training Data
# ============================================================================

# @title Prepare Enhanced Training Data

print("\n" + "="*60)
print("BUILDING ENHANCED FEATURES")
print("="*60)

# Create ID to name mapping
id_to_name = {idx: name for name, idx in neuron_map.items()}

# 1. Build initial connectivity (average L1)
print("\n1. Creating initial state (L1 average)...")
l1_matrices = torch.stack(data_map[0.0])
avg_l1 = torch.mean(l1_matrices, dim=0)

# Get edge index and weights for GNN
edge_index = avg_l1.nonzero().t()
edge_weight = avg_l1[edge_index[0], edge_index[1]]

print(f"   Initial connections: {edge_index.shape[1]}")
print(f"   Initial synapses: {edge_weight.sum().item():.0f}")

# 2. Build comprehensive node features
print("\n2. Building rich node features...")

# a) Identity encoding (one-hot)
identity_features = torch.eye(NUM_NODES)

# b) Cell type encoding (one-hot: sensory, inter, motor, modulatory, muscle)
cell_type_features = torch.zeros(NUM_NODES, 5)
for idx, cell_type in cell_type_dict.items():
    cell_type_features[idx, cell_type] = 1.0

# c) Spatial coordinates (normalized)
spatial_features = torch.zeros(NUM_NODES, 3)
if coord_dict is not None:
    coords_array = []
    for idx in range(NUM_NODES):
        name = id_to_name[idx]
        if name in coord_dict:
            coords_array.append(coord_dict[name])
        else:
            coords_array.append([0, 0, 0])

    coords_array = np.array(coords_array)
    # Normalize coordinates
    coords_mean = coords_array.mean(axis=0)
    coords_std = coords_array.std(axis=0) + 1e-6
    coords_normalized = (coords_array - coords_mean) / coords_std
    spatial_features = torch.tensor(coords_normalized, dtype=torch.float32)

# d) Initial connectivity features
in_degree = avg_l1.sum(dim=0).unsqueeze(1)  # Incoming connections
out_degree = avg_l1.sum(dim=1).unsqueeze(1)  # Outgoing connections
total_degree = (in_degree + out_degree)

# Normalize degrees
max_degree = total_degree.max() + 1e-6
degree_features = torch.cat([
    in_degree / max_degree,
    out_degree / max_degree,
    total_degree / max_degree
], dim=1)

# Combine all features
x_init = torch.cat([
    identity_features,
    cell_type_features,
    spatial_features,
    degree_features
], dim=1).to(device)

print(f"   Feature dimensions: {x_init.shape[1]}")
print(f"     - Identity: {NUM_NODES}")
print(f"     - Cell type: 5")
print(f"     - Spatial: 3")
print(f"     - Degree: 3")

# 3. Prepare contact matrix
contact_matrix_tensor = torch.tensor(contact_matrix, dtype=torch.float32).to(device)
print(f"\n3. Physical contact matrix ready: {contact_matrix_tensor.shape}")

# 4. Prepare targets (all time points, all replicates)
print("\n4. Preparing training targets...")

sorted_times = sorted(data_map.keys())
target_times = torch.tensor(sorted_times).float().to(device)

# Create list of all available ground truth matrices
all_targets = []
all_target_times = []
all_dataset_ids = []

dataset_id = 0
for t in sorted_times:
    for mat in data_map[t]:
        all_targets.append(mat.to(device))
        all_target_times.append(t)
        all_dataset_ids.append(dataset_id)
        dataset_id += 1

print(f"   Total training samples: {len(all_targets)}")
print(f"   Time points: {sorted_times}")
print(f"   Replicates per time: {[len(data_map[t]) for t in sorted_times]}")

# 5. Calculate max synapses for normalization
MAX_SYNAPSES = max([mat.max().item() for mat in all_targets])
print(f"\n5. Max synapses per connection: {MAX_SYNAPSES:.0f}")

print("\n" + "="*60)
print("FEATURE PREPARATION COMPLETE")
print("="*60)
```

```python
# ============================================================================
# BLOCK 4: Define Biologically-Informed Architecture
# ============================================================================

# @title Model Architecture

class BiologicalODEFunc(nn.Module):
    """
    Biologically-informed ODE function
    Separate pathways for different developmental processes
    """
    def __init__(self, hidden_dim):
        super(BiologicalODEFunc, self).__init__()

        # Feedforward enhancement pathway (increases with time)
        self.feedforward_net = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, hidden_dim)
        )

        # Hub amplification pathway (strengthens well-connected neurons)
        self.hub_amplification = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, hidden_dim)
        )

        # Modularity enhancement pathway (increases segregation)
        self.modularity_net = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, hidden_dim)
        )

        # Time-dependent gating (blends processes based on developmental stage)
        self.time_gate = nn.Sequential(
            nn.Linear(1, hidden_dim),
            nn.Sigmoid()
        )

        # Combine pathways
        self.combiner = nn.Linear(hidden_dim * 3, hidden_dim)

    def forward(self, t, z):
        # Time embedding for gating
        t_tensor = t * torch.ones(1, 1, device=z.device)
        time_weight = self.time_gate(t_tensor).squeeze(0)

        # Apply different developmental processes
        dz_ff = self.feedforward_net(z)
        dz_hub = self.hub_amplification(z)
        dz_mod = self.modularity_net(z)

        # Concatenate and combine
        dz_combined = torch.cat([dz_ff, dz_hub, dz_mod], dim=-1)
        dz = self.combiner(dz_combined)

        # Time-dependent weighting (more feedforward/modular with time)
        dz = dz * (0.5 + 0.5 * time_weight)

        return dz


class StochasticODEFunc(nn.Module):
    """
    Add stochasticity for variable connections
    Captures the ~43% variable connections observed in biology
    """
    def __init__(self, deterministic_func, noise_scale=0.01):
        super(StochasticODEFunc, self).__init__()
        self.deterministic_func = deterministic_func
        self.noise_scale = nn.Parameter(torch.tensor(noise_scale))

    def forward(self, t, z):
        # Deterministic component
        dz_det = self.deterministic_func(t, z)

        # Stochastic component (Brownian motion)
        if self.training:
            noise = torch.randn_like(z) * torch.abs(self.noise_scale)
            return dz_det + noise
        else:
            return dz_det


class ImprovedConnectomeODE(nn.Module):
    """
    Complete model incorporating 10 features:
    1. Multi-replicate training
    2. Better loss function (Poisson)
    3. Physical contact scaffold
    4. Biological ODE function
    5. Rich node features
    6. Multi-resolution supervision
    7. Stochastic connections
    8. Biological regularization
    9. Better evaluation
    10. Attention mechanism (GAT)
    """
    def __init__(self, num_nodes, input_dim, hidden_dim=64, heads=4,
                 use_attention=True, use_stochastic=True):
        super(ImprovedConnectomeODE, self).__init__()

        self.num_nodes = num_nodes
        self.hidden_dim = hidden_dim
        self.use_attention = use_attention

        # IMPROVEMENT #10: Graph Attention instead of GCN
        if use_attention:
            self.encoder = GATConv(input_dim, hidden_dim // heads, heads=heads,
                                  dropout=0.1, concat=True)
        else:
            self.encoder = GCNConv(input_dim, hidden_dim)

        # IMPROVEMENT #3: Encode physical contact information
        self.contact_encoder = nn.Sequential(
            nn.Linear(num_nodes, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, hidden_dim // 4)
        )

        # Combined embedding dimension
        combined_dim = hidden_dim + hidden_dim // 4

        # IMPROVEMENT #4: Biological ODE function
        base_ode_func = BiologicalODEFunc(combined_dim)

        # IMPROVEMENT #7: Add stochasticity
        if use_stochastic:
            self.ode_func = StochasticODEFunc(base_ode_func, noise_scale=0.02)
        else:
            self.ode_func = base_ode_func

        # Decoder: Inner product with learned scaling
        self.scale = nn.Parameter(torch.tensor(1.0))
        self.bias = nn.Parameter(torch.zeros(1))

        # Additional decoder layers for better reconstruction
        self.decoder = nn.Sequential(
            nn.Linear(combined_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, combined_dim)
        )

    def forward(self, x, edge_index, edge_weight, contact_matrix, t_span):
        """
        Forward pass with enhanced features

        Args:
            x: Node features [num_nodes, input_dim]
            edge_index: Edge connectivity [2, num_edges]
            edge_weight: Edge weights [num_edges]
            contact_matrix: Physical contacts [num_nodes, num_nodes]
            t_span: Time points to integrate [num_time_points]

        Returns:
            Predicted adjacency matrices at each time point
        """
        # Encode graph structure
        z0_graph = self.encoder(x, edge_index, edge_weight)

        # Encode physical contact information
        z0_contact = self.contact_encoder(contact_matrix)

        # Combine graph and contact embeddings
        z0 = torch.cat([z0_graph, z0_contact], dim=1)

        # Integrate ODE forward in time
        z_t = odeint(self.ode_func, z0, t_span,
                    rtol=1e-3, atol=1e-4, method='dopri5')

        # Decode to adjacency matrices
        predictions = []
        for i in range(len(t_span)):
            z = z_t[i]

            # Additional decoding
            z_decoded = self.decoder(z)

            # Link prediction via inner product
            logits = torch.mm(z_decoded, z_decoded.t()) * self.scale + self.bias

            predictions.append(logits)

        return torch.stack(predictions)

    def get_attention_weights(self, x, edge_index, edge_weight):
        """Extract attention weights for interpretation"""
        if self.use_attention and hasattr(self.encoder, 'lin_l'):
            # Get attention from GAT
            return self.encoder(x, edge_index, edge_weight, return_attention_weights=True)
        return None


# Initialize model
INPUT_DIM = x_init.shape[1]
HIDDEN_DIM = 128
HEADS = 8

model = ImprovedConnectomeODE(
    num_nodes=NUM_NODES,
    input_dim=INPUT_DIM,
    hidden_dim=HIDDEN_DIM,
    heads=HEADS,
    use_attention=True,
    use_stochastic=True
).to(device)

# Move data to device
edge_index = edge_index.to(device)
edge_weight = edge_weight.to(device)

print("\n" + "="*60)
print("MODEL ARCHITECTURE")
print("="*60)
print(f"Total parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"Input dimension: {INPUT_DIM}")
print(f"Hidden dimension: {HIDDEN_DIM}")
print(f"Attention heads: {HEADS}")
print(f"Using GAT: {model.use_attention}")
print("="*60)
```

```python
# ==== PRECOMPUTE FEEDFORWARD MASK ON GPU (ONCE) ====
ff_mask = torch.zeros((NUM_NODES, NUM_NODES), device=device)

for i in range(NUM_NODES):
    for j in range(NUM_NODES):
        ti, tj = cell_type_dict[i], cell_type_dict[j]
        if (ti == 0 and tj in [1, 2]) or \
           (ti == 1 and tj == 2) or \
           (ti in [0, 1, 2] and tj == 4):
            ff_mask[i, j] = 1.0
```

```python
# ======================================================================
# BLOCK 5: Training with Biological Regularization (GPU-OPTIMIZED)
# ======================================================================

def poisson_loss(pred, target):
    pred_positive = F.softplus(pred)
    return F.poisson_nll_loss(pred_positive, target, log_input=False, full=True)

def compute_biological_losses(pred_probs, contact_matrix):
    ff_ratio = (pred_probs * ff_mask).sum() / (pred_probs.sum() + 1e-6)
    feedforward_loss = -ff_ratio

    target_sparsity = 0.07
    sparsity_loss = (pred_probs.mean() - target_sparsity) ** 2

    diff = pred_probs - pred_probs.T
    symmetry_loss = (diff.triu(1) ** 2).mean()

    contact_loss = -(pred_probs * contact_matrix).sum() / (contact_matrix.sum() + 1e-6)

    return feedforward_loss, sparsity_loss, symmetry_loss, contact_loss

def evaluate_connectome(pred, gt, threshold=0.5):
    pred_probs = torch.sigmoid(pred)
    pred_binary = (pred_probs > threshold).float()
    gt_binary = (gt > 0).float()

    pred_flat = pred_binary.cpu().numpy().flatten()
    gt_flat = gt_binary.cpu().numpy().flatten()

    pred_counts = pred_probs.detach().cpu().numpy().flatten()
    gt_counts = gt.cpu().numpy().flatten()

    nonzero_mask = gt_counts > 0
    synapse_corr = pearsonr(pred_counts[nonzero_mask], gt_counts[nonzero_mask])[0] \
        if nonzero_mask.sum() > 0 else 0.0

    metrics = {
        'f1_score': f1_score(gt_flat, pred_flat, zero_division=0),
        'synapse_corr': synapse_corr,
        'sparsity': pred_flat.mean(),
        'gt_sparsity': gt_flat.mean()
    }
    return metrics

EPOCHS = 2000
LR = 0.001
BATCH_SIZE = 4

optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)

WEIGHTS = {
    'reconstruction': 1.0,
    'feedforward': 0.05,
    'sparsity': 0.1,
    'symmetry': 0.02,
    'contact': 0.08
}

history = {
    'total_loss': [],
    'recon_loss': [],
    'bio_losses': [],
    'metrics': []
}

scaler = torch.cuda.amp.GradScaler()

print("\n" + "="*60)
print("STARTING TRAINING")
print("="*60)

for epoch in range(EPOCHS + 1):
    model.train()
    epoch_losses = []

    indices = np.random.permutation(len(all_targets))

    for batch_start in range(0, len(all_targets), BATCH_SIZE):
        batch_indices = indices[batch_start:batch_start + BATCH_SIZE]

        optimizer.zero_grad()
        batch_loss = 0

        for idx in batch_indices:
            gt = all_targets[idx].to(device)
            t_point = torch.tensor([all_target_times[idx]], device=device)

            with torch.cuda.amp.autocast():
                pred_logits = model(
                    x_init, edge_index, edge_weight,
                    contact_matrix_tensor, t_point
                )

                pred_probs = torch.sigmoid(pred_logits[0])

                recon_loss = poisson_loss(pred_logits[0], gt)
                ff, sp, sy, co = compute_biological_losses(
                    pred_probs, contact_matrix_tensor
                )

                total_loss = (
                    WEIGHTS['reconstruction'] * recon_loss +
                    WEIGHTS['feedforward'] * ff +
                    WEIGHTS['sparsity'] * sp +
                    WEIGHTS['symmetry'] * sy +
                    WEIGHTS['contact'] * co
                )

            batch_loss += total_loss

        batch_loss /= len(batch_indices)

        scaler.scale(batch_loss).backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        scaler.step(optimizer)
        scaler.update()

        epoch_losses.append(batch_loss.item())

    scheduler.step()

    avg_loss = np.mean(epoch_losses)
    history['total_loss'].append(avg_loss)

    if epoch % 100 == 0:
        model.eval()
        with torch.no_grad():
            adult_gt = all_targets[-1].to(device)
            adult_time = torch.tensor([3.0], device=device)

            adult_pred = model(
                x_init, edge_index, edge_weight,
                contact_matrix_tensor, adult_time
            )

            metrics = evaluate_connectome(adult_pred[0], adult_gt)
            history['metrics'].append(metrics)

            print(f"\nEpoch {epoch:4d} | Loss: {avg_loss:.4f} | "
                  f"F1: {metrics['f1_score']:.3f} | "
                  f"Corr: {metrics['synapse_corr']:.3f} | "
                  f"LR: {scheduler.get_last_lr()[0]:.6f}")

print("\n" + "="*60)
print("TRAINING COMPLETE")
print("="*60)
```

```python
# ============================================================================
# BLOCK 10: Interactive Query System
# ============================================================================

# @title Query Specific Time Points

def query_connectome(time_point, source_neuron=None, target_neuron=None,
                    threshold=0.5):
    """
    Query the interpolated connectome at any time point

    Args:
        time_point: Float between 0.0 (birth) and 3.0 (adult)
        source_neuron: Optional source neuron name
        target_neuron: Optional target neuron name
        threshold: Minimum connection strength to display
    """
    model.eval()

    t_query = torch.tensor([time_point]).float().to(device)

    with torch.no_grad():
        pred = model(x_init, edge_index, edge_weight,
                    contact_matrix_tensor, t_query)
        adj = torch.sigmoid(pred[0]).cpu().numpy()

    print(f"\n{'='*60}")
    print(f"CONNECTOME AT TIME t={time_point:.2f}")
    print(f"{'='*60}")

    # Overall statistics
    total_connections = (adj > threshold).sum()
    total_synapses = adj.sum() * MAX_SYNAPSES

    print(f"\nNetwork Statistics:")
    print(f"  Total connections (>{threshold}): {total_connections}")
    print(f"  Estimated total synapses: {total_synapses:.0f}")
    print(f"  Network density: {adj.mean():.4f}")

    # Specific query
    if source_neuron is not None:
        if source_neuron in neuron_map:
            src_idx = neuron_map[source_neuron]
            print(f"\nOutputs from {source_neuron}:")

            targets = []
            for j in range(NUM_NODES):
                if adj[src_idx, j] > threshold:
                    targets.append((id_to_name[j], adj[src_idx, j]))

            targets.sort(key=lambda x: x[1], reverse=True)
            for target, strength in targets[:10]:  # Top 10
                print(f"  → {target:<15} strength: {strength:.3f}")
        else:
            print(f"\nNeuron {source_neuron} not found")

    if target_neuron is not None:
        if target_neuron in neuron_map:
            tgt_idx = neuron_map[target_neuron]
            print(f"\nInputs to {target_neuron}:")

            sources = []
            for i in range(NUM_NODES):
                if adj[i, tgt_idx] > threshold:
                    sources.append((id_to_name[i], adj[i, tgt_idx]))

            sources.sort(key=lambda x: x[1], reverse=True)
            for source, strength in sources[:10]:  # Top 10
                print(f"  {source:<15} → strength: {strength:.3f}")
        else:
            print(f"\nNeuron {target_neuron} not found")

    return adj

# Example queries
print("\n" + "="*60)
print("EXAMPLE QUERIES")
print("="*60)

# Query at intermediate time
query_connectome(1.5, source_neuron='AVAL', threshold=0.3)

# Query development of specific connection
print("\n" + "="*60)
print("DEVELOPMENTAL TRAJECTORY OF SPECIFIC CONNECTION")
print("="*60)

source = 'AVAL'
target = 'DA1'

if source in neuron_map and target in neuron_map:
    src_idx = neuron_map[source]
    tgt_idx = neuron_map[target]

    trajectory = []
    for t_idx in range(len(interp_times)):
        strength = pred_probs[t_idx, src_idx, tgt_idx]
        trajectory.append(strength)

    plt.figure(figsize=(10, 5))
    plt.plot(time_array, trajectory, linewidth=3, color='#E63946', marker='o')
    for t in actual_times:
        plt.axvline(t, color='gray', linestyle='--', alpha=0.3)
    plt.xlabel('Developmental Time', fontsize=12)
    plt.ylabel('Connection Strength', fontsize=12)
    plt.title(f'Development of {source} → {target} Connection',
             fontsize=14, fontweight='bold')
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig('specific_connection_trajectory.png', dpi=300)
    plt.show()

    print(f"\n✓ Trajectory plot saved as 'specific_connection_trajectory.png'")
    print(f"\nConnection strength at key time points:")
    for t in actual_times:
        closest_idx = np.argmin(np.abs(time_array - t))
        print(f"  t={t:.1f}: {trajectory[closest_idx]:.4f}")

print("\n" + "="*60)
print("✓ ALL ANALYSES COMPLETE!")
print("="*60)
print("\nGenerated files:")
print("  1. training_analysis.png - Training metrics")
print("  2. connectome_evolution_heatmaps.png - Visual timeline")
print("  3. developmental_principles.png - Biological metrics")
print("  4. specific_connection_trajectory.png - Example connection")
print("  5. connectome_ode_model.pt - Saved model")
print("  6. predicted_connectome_t*.csv - Predictions at each stage")
print("\nYou can now query any time point between 0.0 and 3.0!")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_poincare_disk():
    fig, ax = plt.subplots(figsize=(8, 8))

    # 1. Draw the Boundary (The "Infinity" Circle)
    circle = plt.Circle((0, 0), 1, color='black', fill=False, linewidth=3)
    ax.add_artist(circle)

    # 2. Generate Data Points (Simulating Neurons)
    # Hubs (Center - Low Radius)
    r_hubs = np.random.uniform(0.1, 0.3, 5)
    theta_hubs = np.random.uniform(0, 2*np.pi, 5)
    x_hubs = r_hubs * np.cos(theta_hubs)
    y_hubs = r_hubs * np.sin(theta_hubs)

    # Interneurons (Middle)
    r_inter = np.random.uniform(0.4, 0.7, 15)
    theta_inter = np.random.uniform(0, 2*np.pi, 15)
    x_inter = r_inter * np.cos(theta_inter)
    y_inter = r_inter * np.sin(theta_inter)

    # Sensory/Motor (Edge - High Radius)
    r_sensor = np.random.uniform(0.85, 0.95, 30)
    theta_sensor = np.random.uniform(0, 2*np.pi, 30)
    x_sensor = r_sensor * np.cos(theta_sensor)
    y_sensor = r_sensor * np.sin(theta_sensor)

    # 3. Plot Points
    ax.scatter(x_hubs, y_hubs, c='red', s=200, label='Hubs (AVA, AVE)', zorder=3)
    ax.scatter(x_inter, y_inter, c='blue', s=100, label='Interneurons', zorder=2)
    ax.scatter(x_sensor, y_sensor, c='green', s=50, label='Sensors/Motors', zorder=1)

    # 4. Draw a "Geodesic" (Curved Line)
    # A straight line in Hyperbolic space looks like an arc in the disk
    # Simple visualization: connect a Hub to a Sensor
    t = np.linspace(0, 1, 100)
    # This is a visual approximation of a geodesic arc
    arc_x = np.linspace(x_hubs[0], x_sensor[0], 100)
    bend = 0.2 # curve factor
    arc_y = np.linspace(y_hubs[0], y_sensor[0], 100) + bend * np.sin(np.pi * t)
    ax.plot(arc_x, arc_y, color='gray', linestyle='--', alpha=0.6, label='Geodesic Connection')

    # Formatting
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.title("Connectome Embedding in Poincaré Disk", fontsize=15)
    plt.legend(loc='upper right')
    plt.show()

plot_poincare_disk()
```
